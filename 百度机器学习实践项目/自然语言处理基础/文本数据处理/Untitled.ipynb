```python
import numpy as np
import jieba
```

## one-hot

```python
train_data=open('data2.txt','r',encoding='utf-8')
sentences=[]
for data in train_data:
    sentences.append(data)

#     print(data)
#     print(type(data),data)
```

```python
sentences
```

```python
token_index={}
for sentence in sentences:
    for word in jieba.lcut(sentence):
        if word not in token_index:
            token_index[word]=len(token_index)+1
print(token_index)
```

```python
results=np.zeros((len(sentences),max(token_index.values())+1))
for i,sentence in enumerate(sentences):
    for word in jieba.lcut(sentence):
        j=token_index.get(word)
        results[i,j]=1
        
for i in range(len(sentences)):
    print(f'data:{sentences[i][:-1]}')
    print(f'ont-hot:{results[i]}')
```

## TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
def sklearn_tfidf():
    train_data=open('data1.txt')
    sentences=[]
    for data in train_data:
        sentences.append(data)
    
    vectorizer=CountVectorizer()
    
    X=vectorizer.fit_transform(sentences)
    
    transformer=TfidfTransformer()
    tfidf=transformer.fit_transform(X)
    for I in range(len(sentences)):
        print(sentences[I])
        print(tfidf.toarray()[I])
```

```python
sklearn_tfidf()
```
