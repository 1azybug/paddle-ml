```python
import paddle
import numpy as np
import os
import matplotlib.pyplot as plt 
```

```python
paddle.set_default_dtype('float64')
train_data=paddle.text.datasets.UCIHousing(mode='train')
test_data=paddle.text.datasets.UCIHousing(mode='test')
```

```python
train_it=paddle.io.DataLoader(train_data,batch_size=32,shuffle=True)
test_it=paddle.io.DataLoader(test_data,batch_size=8,shuffle=False)
```


## 配置模型

```python
class Regressor(paddle.nn.Layer):
    def __init__(self):
        super(Regressor,self).__init__()
        self.linear=paddle.nn.Linear(13,1)
    
    def forward(self,inputs):
        x=self.linear(inputs)
        return x
```

## 模型训练

```python
Batch=0
Batchs=[]
losses=[]

model=Regressor()
model.train()
mse=paddle.nn.MSELoss()
opt=paddle.optimizer.SGD(learning_rate=0.0005,parameters=model.parameters())
```

```python
epochs_num=2000
for pass_num in range(epochs_num):
    for batch_id,data in enumerate(train_it()):
        x=data[0]
        y=data[1]
        
        pred=model(x)
        # print(y.numpy())
        loss=mse(pred,y)
        # print(loss)
        # print(loss.numpy())
        if batch_id !=0 and batch_id % 10 ==0:
            Batch+=10
            Batchs.append(Batch)
            losses.append(loss.numpy()[0])
            print("epch:{}, step:{}, loss:{}".format(pass_num,batch_id,loss.numpy()[0]))

        loss.backward() 
        opt.step()   
        opt.clear_grad()
paddle.save(model.state_dict(),"Regressor")
```

## 模型评估

#### loss图

```python
def draw(x,y):
    plt.title('training loss')
    plt.xlabel('batch')
    plt.ylabel('loss')
    plt.plot(x,y,c='g',label="training loss")
    plt.grid()
    plt.legend()
```

```python
draw(Batchs,losses)
```

#### 验证集loss

```python
model=Regressor()
model.set_state_dict(paddle.load("Regressor"))
model.eval()
```

```python
losses=[]
infer_results=[]
ground_truths=[]

for batch_id,data in enumerate(test_it()):
    x=data[0]
    y=data[1]

    pred=model(x)

    ground_truths.extend(y.numpy())
    infer_results.extend(pred.numpy())

    loss=mse(y,pred)
    losses.append(loss.numpy()[0])

print("mean_loss:{}".format(np.mean(losses)))
```

#### y=x绘图

```python
def draw_infer_result(ground_truths,infer_results):
    x=np.arange(1,40)
    y=x
    plt.plot(x,y)
    plt.scatter(ground_truths,infer_results)
```

```python
draw_infer_result(ground_truths,infer_results)
```
