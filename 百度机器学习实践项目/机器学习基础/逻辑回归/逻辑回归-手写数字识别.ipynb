```python
import struct
import matplotlib.pyplot as plt
import os
from array import array as pyarray
import numpy as np
from sklearn.metrics import accuracy_score,classification_report
```

```python
train_img_path=os.path.join("mnist","train-images.idx3-ubyte")
train_lbl_path=os.path.join("mnist","train-labels.idx1-ubyte")
test_img_path=os.path.join("mnist","t10k-images.idx3-ubyte")
test_lbl_path=os.path.join("mnist","t10k-labels.idx1-ubyte")
```

```python
def load_mnist(img_path,lbl_path):
    
    flbl=open(lbl_path,'rb')
    magic_nr,size=struct.unpack(">II",flbl.read(8))
    lbl=pyarray("b",flbl.read())
    flbl.close()
#     print("size=",size)
    
    
    fimg=open(img_path,'rb')
    magic_nr,size,rows,cols=struct.unpack(">IIII",fimg.read(16))
    img=pyarray("B",fimg.read())
    fimg.close()
#     print("size=",size)
    
    digits=np.arange(10)
    ind=[k for k in range(size) if lbl[k] in digits]
    
    N=len(ind)
#     print(N)
    
    imgs=np.zeros((N,rows*cols),dtype=np.uint8)
    lbls=np.zeros((N,1),dtype=np.int8)
    
    for i in range(N):
        imgs[i]=np.array(img[ind[i]*rows*cols:(ind[i]+1)*rows*cols]).reshape(1,rows*cols)
        lbls[i]=lbl[ind[i]]
    
    return imgs,lbls
    
```

```python
train_img,train_lbl=load_mnist(train_img_path,train_lbl_path)
test_img,test_lbl=load_mnist(test_img_path,test_lbl_path)
```

```python
def showim(img,lbl):
#     plt.imshow(img.reshape(28,28),cmap='gray',interpolation='nearest')
#     plt.title('label:%i'%lbl)    
    
    for i,(im,lb) in enumerate(list(zip(img,lbl))):
        plt.subplots_adjust(left=1,right=3,top=2)
        plt.subplot(5,10,i+1)
        plt.axis("off")
        plt.imshow(im.reshape(28,28),cmap='gray',interpolation='nearest')
        plt.title("label=%i"%lb)
    

```

```python
showim(train_img[:50],train_lbl[:50])
```

```python
train_img=train_img.astype(dtype='float32')
test_img=test_img.astype(dtype='float32')
train_img/=255.0
test_img/=255.0
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(train_img,train_lbl)
```

```python
pred=lr.predict(test_img)
print("acc=%.4lf\n"%accuracy_score(pred,test_lbl))
print("Classification report for classifier %s:\n%s\n"%(lr,classification_report(test_lbl,pred)))
```
